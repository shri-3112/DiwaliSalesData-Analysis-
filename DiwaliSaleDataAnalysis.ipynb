# Data Cleaning and prepartion/processing

```python
# drop unrelated/blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.shape #To check if 2 columns are dropped or not
```

```python
# To check whether there are any null values in data
pd.isnull(df)
```

```python
# to check for all null values columnwise
pd.isnull(df).sum()
```

```python
# To drop null values from table
df.dropna(inplace=True)
```

```python
# To check if rows with null values are deleted or no
#df.shape

pd.isnull(df).sum()
```

```python
# To change datatype of column
df['Amount'] = df['Amount'].astype('int')
```

```python
# Tocheck whether data type of Amount column is changed or not.
df['Amount'].dtypes
```

```python
# to check columns in data
df.columns
```

```python
# To get description (i.e.count,mean,std etc) of the data
df.describe()
```

```python
# To replace 0 and 1 values Marital_status column with married/unmarried values
df['Marital_Status'].replace({0:'Married',1:'Unmarried'},inplace=True)
df['Marital_Status'].head()
```

# Emploratory Data Analysis

Gender wise purchases during Diwali

```python
# To find out the genderwise count of orders
ax=sns.countplot(x='Gender',data=df,hue='Gender') 
sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To find out the total amount spent on Diwali purchase genderwise
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue='Gender')
```

From above graphs we can see that most of the buyers are females and even the amount spent on purchsing is more for females then men

Agewise purchase during Diwali


```python
df.columns
```

```python
ax=sns.countplot(x='Age Group', data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To find out Total Amount Vs Age Group
sales_age =df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Age Group')
```

From above graphs we can see that most of the buyers are from 26-35 age group and mostly females

To get statewise Diwali purchase data 

```python
# To find total no of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders',hue='State')
```

```python
# To find amount/sales for top 10 states

states_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=states_amount,x='State',y='Amount',hue='State')
```

From the above charts we can see that most of the orders & total sales/amount are from Uttar Pradesh,Maharashtra and Karnataka respectively.

To find out relation between Diwali sales and marital status

```python
df['Marital_Status'].dtype
#dtype(df['Marital_Status'])
#df['Marital_Status'].replace({0:'Married',1:'Unmarried'},inplace=True)
```

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Marital_Status')

sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To find the total sales/amount spent by married and unmarried group of people.
states_amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=states_amount,x='Marital_Status',y='Amount',hue='Gender')
```

From these graphs we can see that most of the buyers are married(women) with high purchasing power.

To find the relation between occupation and Diwali purchases

```python
ax=sns.countplot(x='Occupation',data=df,hue='Occupation')

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To find the total sales/amount spent by group of people from different occupations.
states_amount=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=states_amount,x='Occupation',y='Amount',hue='Occupation')
```

From the above graphs we can see that most of the buyers are from IT,Healthcare and Aviation sector.

Product category Vs Diwali Sales

```python
# to show no of orders for each product category
ax=sns.countplot(x='Product_Category',data=df,hue='Product_Category')

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To find the total sales/amount spent by people for different product categories.
states_amount=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=states_amount,x='Product_Category',y='Amount',hue='Product_Category')
```

From the above graphs we can see that most of the sold are from Food, clothing and Electronics.

For finding top 10 most sold products

```python
# To find the total sales/amount spent by people for different product categories.
states_amount=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=states_amount,x='Product_ID',y='Orders',hue='Product_ID')
```

# Married women age group 26-35 years from UP,Maharashtra, Karnataka working in IT,Healthcare and Aviation are more likely to buy products form Food, Clothing and Electronics gadgets category
